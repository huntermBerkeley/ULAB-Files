```python
#x = np.loadtxt(open("Data_Entry_2017.csv", "rb"), delimiter=",", skiprows=1)
import csv

name_to_ylabels = {}
with open("Data_Entry_2017.csv") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are ' + ", ".join(row))
        else:
            name_to_ylabels[row[0]] = row[1]
        line_count+=1
```

```python
for i in name_to_ylabels.keys():
    print(name_to_ylabels[i])
```

```python
y_fake = []
for i in name_to_ylabels.keys():
    y_fake.append(name_to_ylabels[i].split("|")[0])
    
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

#enc = OneHotEncoder(handle_unknown='ignore')
le = LabelEncoder()
```

```python
le.fit(y_fake)
```

```python
NUM_CLASSES = len(le.classes_)
```

```python
y_labels = []

for i in total_files:
    y_label = [0]*NUM_CLASSES
    i_index = i.split("/")[-1]
    for descriptor in name_to_ylabels[i_index].split("|"):
        #print(descriptor)
        #print(le.transform([descriptor]))
        y_label[le.transform([descriptor])[0]] = 1
    y_labels.append(y_label)
        
    #row = labels.loc[labels["Image Index"] == i.split("/")[-1]]
  
```

```python
y_labels = np.asarray(y_labels)
```
